```python
# Written by Ovi
# Summary: Comprehensive analysis for multiple CSV datasets with advanced EDA, visualizations, and comparisons.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import warnings
warnings.filterwarnings('ignore')

# Set up plotting style
plt.style.use('default')
sns.set_palette("husl")

# List of dataset file paths
file_paths = [
    "dataset1.csv",
    "dataset2.csv", 
    "dataset3.csv",
    "dataset4.csv",
    "dataset5.csv",
    "dataset6.csv"
]

# Storage for comparative analysis
dataset_summaries = {}
all_datasets = {}

def safe_read_csv(path):
    """Safely read CSV with different encodings"""
    encodings = ['utf-8', 'latin-1', 'iso-8859-1', 'cp1252']
    
    for encoding in encodings:
        try:
            df = pd.read_csv(path, encoding=encoding)
            print(f"✓ Successfully read {path} with {encoding} encoding")
            return df
        except UnicodeDecodeError:
            continue
        except Exception as e:
            print(f"Error reading {path} with {encoding}: {e}")
            continue
    
    raise Exception(f"Could not read {path} with any encoding")

def classify_features(df):
    """Intelligently classify features into binary, categorical, and numerical"""
    
    # Exclude obvious non-feature columns
    exclude_cols = []
    for col in df.columns:
        col_lower = col.lower().strip()
        if any(keyword in col_lower for keyword in ['url', 'filename', 'domain', 'title', 'id', 'index']):
            # Check if it's actually an ID column (sequential numbers)
            if df[col].dtype in ['object', 'string'] or (df[col].nunique() == len(df)):
                exclude_cols.append(col)
    
    feature_cols = [col for col in df.columns if col not in exclude_cols]
    
    binary_features = []
    categorical_features = []
    numerical_features = []
    
    for col in feature_cols:
        try:
            # Convert to numeric if possible
            if df[col].dtype == 'object':
                try:
                    df[col] = pd.to_numeric(df[col], errors='coerce')
                except:
                    pass
            
            # Skip if still non-numeric
            if df[col].dtype == 'object':
                continue
                
            unique_vals = df[col].nunique()
            unique_set = set(df[col].dropna().unique())
            
            # Binary: exactly 2 unique values or subset of {-1, 0, 1}
            if unique_vals == 2 or unique_set.issubset({-1, 0, 1}):
                binary_features.append(col)
            # Categorical: 3-10 unique values
            elif 3 <= unique_vals <= 10:
                categorical_features.append(col)
            # Numerical: more than 10 unique values
            else:
                numerical_features.append(col)
                
        except Exception as e:
            print(f"Warning: Could not classify column {col}: {e}")
            continue
    
    return binary_features, categorical_features, numerical_features, exclude_cols

def safe_outlier_detection(df, numerical_features):
    """Safely detect outliers only for truly numerical columns"""
    outlier_summary = {}
    
    for col in numerical_features:
        try:
            # Ensure column is numeric and has enough variation
            if df[col].dtype in ['object', 'string']:
                continue
                
            numeric_col = pd.to_numeric(df[col], errors='coerce')
            if numeric_col.isna().all() or numeric_col.nunique() < 3:
                continue
                
            Q1 = numeric_col.quantile(0.25)
            Q3 = numeric_col.quantile(0.75)
            IQR = Q3 - Q1
            
            if IQR == 0:  # No variation
                outlier_summary[col] = 0
                continue
                
            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR
            outliers = numeric_col[(numeric_col < lower_bound) | (numeric_col > upper_bound)]
            outlier_summary[col] = len(outliers.dropna())
            
        except Exception as e:
            print(f"Warning: Could not detect outliers for {col}: {e}")
            outlier_summary[col] = 0
            
    return outlier_summary

def safe_correlation_analysis(df, feature_cols):
    """Safely perform correlation analysis"""
    try:
        # Select only numeric columns for correlation
        numeric_df = df[feature_cols].select_dtypes(include=[np.number])
        if numeric_df.shape[1] < 2:
            return pd.DataFrame(), []
            
        correlation_matrix = numeric_df.corr()
        
        # Find highly correlated features
        high_corr = []
        for i in range(len(correlation_matrix.columns)):
            for j in range(i+1, len(correlation_matrix.columns)):
                corr_val = correlation_matrix.iloc[i, j]
                if not np.isnan(corr_val) and abs(corr_val) > 0.8:
                    high_corr.append({
                        'Feature1': correlation_matrix.columns[i],
                        'Feature2': correlation_matrix.columns[j],
                        'Correlation': corr_val
                    })
        
        return correlation_matrix, high_corr
    except Exception as e:
        print(f"Warning: Correlation analysis failed: {e}")
        return pd.DataFrame(), []

def safe_visualizations(df, dataset_name, binary_features, categorical_features, numerical_features, correlation_matrix, high_corr, outlier_summary):
    """Create visualizations with proper error handling"""
    
    try:
        # Create subplots
        fig = plt.figure(figsize=(20, 16))
        
        # Plot 1: Target distribution (look for common target names)
        target_col = None
        for col in ['Result', 'label', 'status', 'CLASS_LABEL', 'phishing']:
            if col in df.columns:
                target_col = col
                break
        
        plt.subplot(3, 3, 1)
        if target_col:
            target_counts = df[target_col].value_counts()
            # Handle negative values for pie chart
            if (target_counts < 0).any():
                # Use bar chart instead of pie for negative values
                target_counts.plot(kind='bar', color=['skyblue', 'lightcoral'])
                plt.title(f'Target Distribution - {dataset_name}')
                plt.ylabel('Count')
                plt.xticks(rotation=0)
            else:
                plt.pie(target_counts.values, labels=target_counts.index, autopct='%1.1f%%', colors=['skyblue', 'lightcoral'])
                plt.title(f'Target Distribution - {dataset_name}')
        else:
            plt.text(0.5, 0.5, 'No Target Column Found', ha='center', va='center', fontsize=14)
            plt.title('Target Distribution')
        
        # Plot 2: Missing data heatmap
        plt.subplot(3, 3, 2)
        missing_data = df.isnull().sum()
        if missing_data.sum() > 0:
            # Show only columns with missing data
            missing_cols = missing_data[missing_data > 0].index[:20]  # Limit to 20 columns
            if len(missing_cols) > 0:
                sns.heatmap(df[missing_cols].isnull(), cbar=True, cmap='viridis')
                plt.title('Missing Data Pattern')
            else:
                plt.text(0.5, 0.5, 'No Missing Data', ha='center', va='center', fontsize=14)
                plt.title('Missing Data Pattern')
        else:
            plt.text(0.5, 0.5, 'No Missing Data', ha='center', va='center', fontsize=14)
            plt.title('Missing Data Pattern')
        
        # Plot 3: Feature distribution for numerical features
        plt.subplot(3, 3, 3)
        if numerical_features:
            sample_features = numerical_features[:4]
            for i, col in enumerate(sample_features):
                try:
                    numeric_col = pd.to_numeric(df[col], errors='coerce').dropna()
                    if len(numeric_col) > 0:
                        plt.hist(numeric_col, alpha=0.7, label=col[:15], bins=20)
                except:
                    continue
            plt.title('Sample Numerical Features')
            plt.legend()
        else:
            plt.text(0.5, 0.5, 'No Numerical Features', ha='center', va='center', fontsize=14)
            plt.title('Sample Numerical Features')
        
        # Plot 4: Correlation heatmap (top features)
        plt.subplot(3, 3, 4)
        if not correlation_matrix.empty and correlation_matrix.shape[0] > 1:
            # Select top correlated features (limit to 15 for readability)
            corr_sum = correlation_matrix.abs().sum().nlargest(15)
            top_features = corr_sum.index
            mini_corr = correlation_matrix.loc[top_features, top_features]
            sns.heatmap(mini_corr, annot=True, cmap='coolwarm', center=0, fmt='.2f', cbar=True)
            plt.title('Top Features Correlation')
        else:
            plt.text(0.5, 0.5, 'Insufficient Data for Correlation', ha='center', va='center', fontsize=12)
            plt.title('Features Correlation')
        
        # Plot 5: Outlier detection visualization
        plt.subplot(3, 3, 5)
        if numerical_features and outlier_summary:
            # Find feature with most outliers for visualization
            if outlier_summary:
                max_outlier_feature = max(outlier_summary, key=outlier_summary.get)
                if outlier_summary[max_outlier_feature] > 0:
                    try:
                        numeric_col = pd.to_numeric(df[max_outlier_feature], errors='coerce').dropna()
                        plt.boxplot(numeric_col)
                        plt.title(f'Outliers: {max_outlier_feature[:20]}')
                        plt.ylabel(max_outlier_feature[:20])
                    except:
                        plt.text(0.5, 0.5, 'Cannot visualize outliers', ha='center', va='center', fontsize=12)
                        plt.title('Outlier Analysis')
                else:
                    plt.text(0.5, 0.5, 'No outliers detected', ha='center', va='center', fontsize=12)
                    plt.title('Outlier Analysis')
            else:
                plt.text(0.5, 0.5, 'No outlier data', ha='center', va='center', fontsize=12)
                plt.title('Outlier Analysis')
        else:
            plt.text(0.5, 0.5, 'No Numerical Features', ha='center', va='center', fontsize=12)
            plt.title('Outlier Analysis')
        
        # Plot 6: Feature importance (based on correlation with target)
        plt.subplot(3, 3, 6)
        if target_col and not correlation_matrix.empty and target_col in correlation_matrix.columns:
            try:
                feature_importance = correlation_matrix[target_col].abs().sort_values(ascending=False)[1:11]
                if len(feature_importance) > 0:
                    feature_importance.plot(kind='barh', color='lightgreen')
                    plt.title('Top Features by Correlation with Target')
                else:
                    plt.text(0.5, 0.5, 'No feature correlations', ha='center', va='center', fontsize=12)
                    plt.title('Feature Importance')
            except:
                plt.text(0.5, 0.5, 'Cannot calculate feature importance', ha='center', va='center', fontsize=12)
                plt.title('Feature Importance')
        else:
            plt.text(0.5, 0.5, 'No Target Column or Correlations', ha='center', va='center', fontsize=12)
            plt.title('Feature Importance')
        
        # Plot 7: Data types distribution
        plt.subplot(3, 3, 7)
        try:
            dtype_counts = df.dtypes.value_counts()
            colors = plt.cm.Set3(np.linspace(0, 1, len(dtype_counts)))
            plt.pie(dtype_counts.values, labels=[str(x) for x in dtype_counts.index], autopct='%1.1f%%', colors=colors)
            plt.title('Data Types Distribution')
        except:
            plt.text(0.5, 0.5, 'Cannot show data types', ha='center', va='center', fontsize=12)
            plt.title('Data Types Distribution')
        
        # Plot 8: Feature types distribution
        plt.subplot(3, 3, 8)
        feature_type_counts = {
            'Binary': len(binary_features),
            'Categorical': len(categorical_features), 
            'Numerical': len(numerical_features)
        }
        if sum(feature_type_counts.values()) > 0:
            plt.pie(feature_type_counts.values(), labels=feature_type_counts.keys(), 
                   autopct='%1.1f%%', colors=['lightblue', 'lightgreen', 'lightcoral'])
            plt.title('Feature Types Distribution')
        else:
            plt.text(0.5, 0.5, 'No features classified', ha='center', va='center', fontsize=12)
            plt.title('Feature Types Distribution')
        
        # Plot 9: Dataset shape info
        plt.subplot(3, 3, 9)
        shape_info = [df.shape[0], df.shape[1]]
        plt.bar(['Rows', 'Columns'], shape_info, color=['orange', 'purple'])
        plt.title(f'Dataset Dimensions\n{df.shape[0]:,} rows × {df.shape[1]} columns')
        plt.ylabel('Count')
        
        plt.tight_layout()
        plt.savefig(f'{dataset_name}_comprehensive_analysis.png', dpi=300, bbox_inches='tight')
        plt.show()
        
    except Exception as e:
        print(f"Error creating visualizations: {e}")

def comprehensive_analysis(path):
    """Perform comprehensive analysis on a single dataset"""
    try:
        # Load dataset with safe encoding
        df = safe_read_csv(path)
        dataset_name = path.split('.')[0]
        all_datasets[dataset_name] = df
        
        print(f"\n{'='*60}")
        print(f"COMPREHENSIVE ANALYSIS FOR {path.upper()}")
        print(f"{'='*60}")
        
        # 1. BASIC INFORMATION
        print("\n1. DATASET OVERVIEW")
        print("-" * 30)
        print(f"Shape: {df.shape}")
        print(f"Memory usage: {df.memory_usage(deep=True).sum() / 1024**2:.2f} MB")
        
        # 2. FEATURE CLASSIFICATION
        binary_features, categorical_features, numerical_features, exclude_cols = classify_features(df)
        
        print("\n2. FEATURE CLASSIFICATION")
        print("-" * 30)
        print(f"Excluded columns ({len(exclude_cols)}): {exclude_cols[:5]}{'...' if len(exclude_cols) > 5 else ''}")
        print(f"Binary features ({len(binary_features)}): {binary_features[:5]}{'...' if len(binary_features) > 5 else ''}")
        print(f"Categorical features ({len(categorical_features)}): {categorical_features[:5]}{'...' if len(categorical_features) > 5 else ''}")
        print(f"Numerical features ({len(numerical_features)}): {numerical_features[:5]}{'...' if len(numerical_features) > 5 else ''}")
        
        # 3. DATA QUALITY ASSESSMENT
        print("\n3. DATA QUALITY")
        print("-" * 30)
        missing_data = df.isnull().sum()
        missing_percent = (missing_data / len(df)) * 100
        
        if missing_data.sum() == 0:
            print("✓ No missing values found!")
        else:
            missing_summary = pd.DataFrame({
                'Missing_Count': missing_data[missing_data > 0],
                'Missing_Percent': missing_percent[missing_data > 0]
            }).sort_values('Missing_Count', ascending=False)
            print("Missing data summary:")
            print(missing_summary)
        
        # 4. TARGET VARIABLE ANALYSIS
        target_col = None
        for col in ['Result', 'label', 'status', 'CLASS_LABEL', 'phishing']:
            if col in df.columns:
                target_col = col
                break
                
        if target_col:
            print(f"\n4. TARGET VARIABLE ANALYSIS ({target_col})")
            print("-" * 30)
            target_dist = df[target_col].value_counts()
            print("Target distribution:")
            print(target_dist)
            if len(target_dist) >= 2:
                balance_ratio = target_dist.min() / target_dist.max()
                print(f"Class balance ratio: {balance_ratio:.3f}")
                if balance_ratio < 0.3:
                    print("⚠️ Significant class imbalance detected!")
            
        # 5. STATISTICAL SUMMARY
        print("\n5. STATISTICAL SUMMARY")
        print("-" * 30)
        try:
            # Only show stats for first 10 numeric columns to avoid clutter
            numeric_cols = df.select_dtypes(include=[np.number]).columns[:10]
            if len(numeric_cols) > 0:
                print(df[numeric_cols].describe().round(3))
            else:
                print("No numerical columns for statistical summary")
        except Exception as e:
            print(f"Could not generate statistical summary: {e}")
        
        # 6. OUTLIER DETECTION
        print("\n6. OUTLIER ANALYSIS")
        print("-" * 30)
        outlier_summary = safe_outlier_detection(df, numerical_features)
        
        if outlier_summary:
            outlier_df = pd.DataFrame(list(outlier_summary.items()), columns=['Feature', 'Outlier_Count'])
            outlier_df['Outlier_Percent'] = (outlier_df['Outlier_Count'] / len(df)) * 100
            outlier_summary_sorted = outlier_df.sort_values('Outlier_Count', ascending=False)
            print(outlier_summary_sorted.head(15))  # Show top 15
        else:
            print("No outlier analysis available")
        
        # 7. CORRELATION ANALYSIS
        print("\n7. CORRELATION ANALYSIS")
        print("-" * 30)
        feature_cols = binary_features + categorical_features + numerical_features
        correlation_matrix, high_corr = safe_correlation_analysis(df, feature_cols)
        
        if high_corr:
            print("Highly correlated feature pairs (|correlation| > 0.8):")
            for pair in high_corr[:10]:  # Show top 10
                print(f"  {pair['Feature1']} <-> {pair['Feature2']}: {pair['Correlation']:.3f}")
        else:
            print("No highly correlated feature pairs found or insufficient data.")
        
        # 8. GENERATE VISUALIZATIONS
        print("\n8. GENERATING VISUALIZATIONS")
        print("-" * 30)
        safe_visualizations(df, dataset_name, binary_features, categorical_features, 
                          numerical_features, correlation_matrix, high_corr, outlier_summary)
        
        # Store summary for comparison
        dataset_summaries[dataset_name] = {
            'shape': df.shape,
            'missing_percent': missing_percent.sum(),
            'binary_features': len(binary_features),
            'categorical_features': len(categorical_features),
            'numerical_features': len(numerical_features),
            'target_balance': target_dist.min()/target_dist.max() if target_col and len(target_dist) >= 2 else None,
            'high_correlations': len(high_corr),
            'total_outliers': sum(outlier_summary.values()) if outlier_summary else 0
        }
        
        print(f"\n✓ Analysis completed for {dataset_name}")
        print(f"✓ Visualization saved as '{dataset_name}_comprehensive_analysis.png'")
        
        return df
        
    except Exception as e:
        print(f"\n❌ Error analyzing {path}: {e}")
        import traceback
        traceback.print_exc()
        return None

def comparative_analysis():
    """Compare all datasets side by side"""
    if len(dataset_summaries) < 2:
        print("Need at least 2 datasets for comparison")
        return
        
    print(f"\n{'='*60}")
    print("COMPARATIVE ANALYSIS ACROSS ALL DATASETS")
    print(f"{'='*60}")
    
    # Create comparison DataFrame
    comparison_df = pd.DataFrame(dataset_summaries).T
    print("\nDataset Comparison Summary:")
    print(comparison_df.round(3))
    
    # Visualization for comparison
    try:
        fig, axes = plt.subplots(2, 3, figsize=(18, 12))
        
        # Shape comparison
        axes[0, 0].bar(comparison_df.index, comparison_df['shape'].apply(lambda x: x[0]))
        axes[0, 0].set_title('Number of Rows Comparison')
        axes[0, 0].tick_params(axis='x', rotation=45)
        
        axes[0, 1].bar(comparison_df.index, comparison_df['shape'].apply(lambda x: x[1]))
        axes[0, 1].set_title('Number of Columns Comparison')
        axes[0, 1].tick_params(axis='x', rotation=45)
        
        # Feature types comparison
        axes[0, 2].bar(comparison_df.index, comparison_df['binary_features'])
        axes[0, 2].set_title('Binary Features Count')
        axes[0, 2].tick_params(axis='x', rotation=45)
        
        axes[1, 0].bar(comparison_df.index, comparison_df['numerical_features'])
        axes[1, 0].set_title('Numerical Features Count')
        axes[1, 0].tick_params(axis='x', rotation=45)
        
        # Data quality comparison
        axes[1, 1].bar(comparison_df.index, comparison_df['missing_percent'])
        axes[1, 1].set_title('Missing Data Percentage')
        axes[1, 1].tick_params(axis='x', rotation=45)
        
        # Outliers comparison
        axes[1, 2].bar(comparison_df.index, comparison_df['total_outliers'])
        axes[1, 2].set_title('Total Outliers Count')
        axes[1, 2].tick_params(axis='x', rotation=45)
        
        plt.tight_layout()
        plt.savefig('datasets_comparison.png', dpi=300, bbox_inches='tight')
        plt.show()
        
        # Save comparison summary to CSV
        comparison_df.to_csv('datasets_comparison_summary.csv')
        print("\n✓ Comparison summary saved to 'datasets_comparison_summary.csv'")
        
    except Exception as e:
        print(f"Error in comparative analysis: {e}")

# MAIN EXECUTION
print("Starting comprehensive analysis of all datasets...")

# Run individual analyses
for path in file_paths:
    result = comprehensive_analysis(path)

# Run comparative analyses
if len(dataset_summaries) > 1:
    comparative_analysis()

print(f"\n{'='*60}")
print("ANALYSIS COMPLETE!")
print(f"{'='*60}")
print("Generated files:")
print("- Individual analysis plots: [dataset_name]_comprehensive_analysis.png")
if len(dataset_summaries) > 1:
    print("- Comparison plot: datasets_comparison.png")
    print("- Comparison summary: datasets_comparison_summary.csv")
```
